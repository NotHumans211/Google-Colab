#Stress Level Classification
Build a classification model to predict stress levels.

```
#Prepare Data
X = df.drop('stress_level', axis=1)
y = df['stress_level']

# Convert categorical variables to dummy variables if needed
X = pd.get_dummies(X, drop_first=True)
```

```
#Split Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
#Train a Model:
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```
#Evaluate the Model:
from sklearn.metrics import classification_report, confusion_matrix
# Import the necessary function
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#Clustering Analysis
Find patterns or groups in the data related to stress levels.

#Perform Clustering

```
import numpy as np  # Import NumPy and alias it as 'np'

from sklearn.cluster import KMeans

# Use only numeric features for clustering
X_cluster = df.select_dtypes(include=[np.number]).drop('stress_level', axis=1)

kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters
clusters = kmeans.fit_predict(X_cluster)

df['Cluster'] = clusters
```

#Visualize Clusters

```
sns.scatterplot(x=df['anxiety_level'], y=df['self_esteem'], hue=df['Cluster'], palette='viridis')
plt.show()
```

# Principal Component Analysis (PCA)
Reduce dimensionality for visualization and better understanding.

#Apply PCA


```
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(X_cluster)

plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['stress_level'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Stress Levels')
plt.colorbar(label='Stress Level')
plt.show()
```
