```python
# Buat DataFrame contoh (ganti dengan data aslimu)
data = {
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'age': [20, 21, 22, 19, 20],
    'university': ['PU', 'PU', 'Other', 'Other', 'PU'],
    'degree_level': ['Undergraduate', 'Undergraduate', 'Graduate', 'Undergraduate', 'Graduate'],
    'degree_major': ['Data Science', 'Engineering', 'Business', 'Psychology', 'Computer Science'],
    'academic_year': ['2nd year', '3rd year', '1st year', '4th year', '2nd year'],
    'cgpa': [3.2, 3.8, 3.5, 3.0, 3.7],
    'residential_status': ['Off-Campus', 'On-Campus', 'Off-Campus', 'On-Campus', 'Off-Campus'],
    'campus_discrimination': ['No', 'Yes', 'No', 'No', 'Yes'],
    'sports_engagement': ['No Sports', 'Regular Sports', 'No Sports', 'Occasional Sports', 'Regular Sports'],
    'average_sleep': ['4-6 hrs', '6-8 hrs', '4-6 hrs', '<4 hrs', '6-8 hrs'],
    'study_satisfaction': [5, 4, 3, 5, 4],
    'academic_workload': [4, 5, 3, 4, 5],
    'academic_pressure': [5, 4, 3, 5, 4],
    'financial_concerns': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'social_relationships': [4, 5, 3, 4, 5],
    'depression': [3, 2, 4, 3, 2],
    'anxiety': [4, 3, 5, 4, 3],
    'isolation': [3, 2, 4, 3, 2],
    'future_insecurity': [4, 3, 5, 4, 3],
    'stress_relief_activities': ['Exercise', 'Socializing', 'Music', 'Reading', 'Gaming']
}
```

```python
df = pd.DataFrame(data)
```

#Distribution of Degree Levels
This will show a bar chart comparing the number of undergraduate and graduate students.

```python
sns.countplot(x='degree_level', data=df)
plt.xlabel('Degree Level')
plt.ylabel('Number of Students')
plt.title('Distribution of Degree Levels')
plt.show()
```

#Relationship between Age and CGPA
This will create a scatter plot showing if there's any correlation between age and CGPA, with different colors representing male and female students.

```python
sns.scatterplot(x='age', y='cgpa', hue='gender', data=df)
plt.xlabel('Age')
plt.ylabel('CGPA')
plt.title('Relationship between Age and CGPA')
plt.show()
```

# Histogram for Depression Distribution

This will show the frequency of each depression level, giving you an idea of how prevalent different levels of depression are.

```python
plt.hist(df['depression'])
plt.xlabel('Depression Level')
plt.ylabel('Frequency')
plt.title('Distribution of Depression Levels')
plt.show()
```

# Box Plot for Depression by Gender
This will compare the distribution of depression levels between males and females, showing potential differences in median, spread, and outliers.

```python
sns.boxplot(x='gender', y='depression', data=df)
plt.xlabel('Gender')
plt.ylabel('Depression Level')
plt.title('Depression Levels by Gender')
plt.show()
```

#Violin Plot for Depression by University
This will provide a more detailed view of depression distribution within each university, showing density and potential differences in shape

```python
sns.violinplot(x='university', y='depression', data=df)
plt.xlabel('University')
plt.ylabel('Depression Level')
plt.title('Depression Levels by University')
plt.show()
```

#Analysis

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Prepare data for the wireframe plot
X = data['academic_pressure']
Y = data['anxiety']
Z = data['study_satisfaction']

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot wireframe
ax.plot_trisurf(X, Y, Z, cmap='viridis', edgecolor='none')

# Label the axes
ax.set_xlabel('Academic Pressure')
ax.set_ylabel('Anxiety')
ax.set_zlabel('Study Satisfaction')

# Title
ax.set_title('3D Visualization: Academic Pressure vs. Anxiety vs. Study Satisfaction')

plt.show()
```

```python
import seaborn as sns
import pandas as pd # Import pandas

# Convert the dictionary to a DataFrame
data = pd.DataFrame(data)

# Calculate the average levels of depression, anxiety, and isolation by university
avg_values = data.groupby('university')[['depression', 'anxiety', 'isolation']].mean().reset_index()

# Melt the data for easier plotting with seaborn
melted_data = avg_values.melt(id_vars='university', var_name='Mental Health Metric', value_name='Average Level')

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='university', y='Average Level', hue='Mental Health Metric', data=melted_data)

# Title and labels
plt.title('Average Levels of Depression, Anxiety, and Isolation by University')
plt.xlabel('University')
plt.ylabel('Average Level')
plt.legend(title='Mental Health Metric')

plt.show()
```
