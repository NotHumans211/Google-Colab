```python
import pandas as pd
from scipy.stats import chi2_contingency

# Print unique values of relevant columns to check for mismatches
print(data['Jenis_Kelamin'].unique())
print(data['Kesulitan_Melepaskan_Diri'].unique())

# Ensure data types are consistent and convert to strings
data['Jenis_Kelamin'] = data['Jenis_Kelamin'].astype(str)
data['Kesulitan_Melepaskan_Diri'] = data['Kesulitan_Melepaskan_Diri'].astype(str)

# Membuat tabel kontingensi antara Jenis_Kelamin dan Kesulitan_Melepaskan_Diri
contingency_table = pd.crosstab(data['Jenis_Kelamin'], data['Kesulitan_Melepaskan_Diri'])

print(contingency_table)

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square: {chi2}, P-value: {p}")
```

# Classification Analysis

Klasifikasi Menggunakan Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Definisikan fitur dan target
X_class = data[['Waktu_Harian', 'Fomo', 'Tidak_Bisa_Kontrol_Diri', 'Mengganggu_Produktivitas']]
y_class = data['Pernah_Memakai_Aplikasi_Pengaturan_Waktu']

# Split data
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Model Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_class, y_train_class)

# Prediksi
y_pred_class = dt_model.predict(X_test_class)

# Evaluasi
print(confusion_matrix(y_test_class, y_pred_class))
print(classification_report(y_test_class, y_pred_class))
```

#Chi-Square Test for Categorical Variables

```python
from scipy.stats import chi2_contingency

# Buat tabel kontingensi
contingency_table = pd.crosstab(data['Jenis_Kelamin'], data['Kesulitan_Melepaskan_Diri'])

# Uji Chi-Square
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}, p-value: {p}")

# Kesimpulan
if p < 0.05:
    print("Terdapat hubungan antara jenis kelamin dan kesulitan melepaskan diri.")
else:
    print("Tidak terdapat hubungan antara jenis kelamin dan kesulitan melepaskan diri.")
```
